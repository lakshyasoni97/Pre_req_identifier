```python
def perform_auto_get_keys(data, level):
    results = []
    iterator = 0

    while True:
        data_n, keys = get_keys(data, level, iterator)
        if data_n is None:
            break

        results.append((data_n, keys))
        data = data_n
        level = keys
        iterator += 1

    return results

data_0, level_0 = data, list(data.keys())
results = perform_auto_get_keys(data, level_0)

for i, result in enumerate(results):
    data_n, keys = result
    globals()[f"data_{i+1}"] = data_n
    globals()[f"level_{i+1}"] = keys

# Access the stored results
print(data_0)
print(level_0)
print(data_1)
print(level_1)
print(data_2)
```
